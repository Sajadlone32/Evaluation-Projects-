# 8.SAVING FEATURES INTO X AN Y VARIABLE

```python
x=df.drop('W',axis=1)
y=df['W']
```

# 9.spliting the data into training and testing datasets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

# 10.IMPORT THE MODELS

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)

svr=SVR()
svr.fit(x_train,y_train)

rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)

gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)

xgb=XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
y_pred2=svr.predict(x_test)
y_pred3=rfr.predict(x_test)
y_pred4=gbr.predict(x_test)
y_pred5=xgb.predict(x_test)
```

# Accuracy Score

```python
from sklearn import metrics
```

```python
#calculating r2 score 
score1=metrics.r2_score(y_test,y_pred1)
score2=metrics.r2_score(y_test,y_pred2)
score3=metrics.r2_score(y_test,y_pred3)
score4=metrics.r2_score(y_test,y_pred4)
score5=metrics.r2_score(y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
final_data=pd.DataFrame({'Models':['LR','SVR','RFR','GRR','XGB'],
             'R2_SCORE':[score1,score2,score3,score4,score5]})
```

```python
final_data
```

```python
sns.barplot(x='Models',y='R2_SCORE',data=final_data)
```
