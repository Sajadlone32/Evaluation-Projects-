# Encoding categorical column

```python
from sklearn.preprocessing import OrdinalEncoder
OE=OrdinalEncoder()
for i in df.columns:
    if df[i].dtype=='object':
        df[i]=OE.fit_transform(df[i].values.reshape(-1,1))
```

```python
df.head()
```

```python
df.info()
```

# droping unneccessary columns

```python
df=df.drop(['Country Code','Restaurant ID'],axis=1)
```

```python
df.columns
```

# Seperating Dependent and independent variables

```python
x=df.drop('Average Cost for two',axis=1)
y=df['Average Cost for two']
```

```python
x
```

# Feature scalling using standard scaller

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

```python
x
```

# Checking vairance inflation factor VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['vif values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

# Import train Test Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

# Importing the models

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor,BaggingRegressor,ExtraTreesRegressor
```

# Training the models

```python
lr=LinearRegression()
lr.fit(x_train,y_train)

svr=SVR()
svr.fit(x_train,y_train)

rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)

gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)

abr=AdaBoostRegressor()
abr.fit(x_train,y_train)

bgr=BaggingRegressor()
bgr.fit(x_train,y_train)

etr=ExtraTreesRegressor()
etr.fit(x_train,y_train)
```

# predicting the data

```python
y_pred1=lr.predict(x_test)
y_pred2=svr.predict(x_test)
y_pred3=rfr.predict(x_test)
y_pred4=gbr.predict(x_test)
y_pred5=abr.predict(x_test)
y_pred6=bgr.predict(x_test)
y_pred7=etr.predict(x_test)
```

# Printing the result 

```python
models=pd.DataFrame({'Actual':y_test,'lr':y_pred1,'svr':y_pred2,'rfr':y_pred3,'gbr':y_pred4,'abr':y_pred5,'bgr':y_pred6,'etr':y_pred7})
models
```

# Checking mean squared error

```python
from sklearn.metrics import mean_squared_error
```

```python
#calculating mean squared error
score1=mean_squared_error(y_test,y_pred1)
score2=mean_squared_error(y_test,y_pred2)
score3=mean_squared_error(y_test,y_pred3)
score4=mean_squared_error(y_test,y_pred4)
score5=mean_squared_error(y_test,y_pred5)
score6=mean_squared_error(y_test,y_pred6)
score7=mean_squared_error(y_test,y_pred7)

```

```python
print(score1,score2,score3,score4,score5,score6,score7)
```
