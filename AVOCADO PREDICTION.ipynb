# Encoding the region column using ordinalEncoder

```python
from sklearn.preprocessing import OrdinalEncoder
enc=OrdinalEncoder()
```

```python
df2=df.copy()
```

```python
df2['region'].value_counts()
```

```python
rs=df2['region'].values.reshape(-1,1)
df2['region']=enc.fit_transform(rs)
df2.head()
```

```python
df2=df2.drop('Date',axis=1)
```

# till now i have completed the EDA and Data preprocessing, now my next step is to seperate dependent and independent variables, in this dataset our Average price is target variable.

# Task 1(target fuature average price)

# 11.Seperating dependent and independent variables into x and y

```python
x=df2.drop('AveragePrice',axis=1)
y=df2['AveragePrice']
```

```python
x['region'].value_counts()
```

# 12.Train Test Split 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

# 13.import the models

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)

svr=SVR()
svr.fit(x_train,y_train)

rfg=RandomForestRegressor()
rfg.fit(x_train,y_train)

gbr=GradientBoostingRegressor()
gbr.fit(x_train,y_train)

```

```python
y_pred1=lr.predict(x_test)
y_pred2=svr.predict(x_test)
y_pred3=rfg.predict(x_test)
y_pred4=gbr.predict(x_test)
```

```python
df1=pd.DataFrame({'Actual':y_test,'lr':y_pred1,'svr':y_pred2,'rfg':y_pred3,'gbr':y_pred4})
```

```python
df1
```

# COMPARE PERFORMANCE VISUALLY

```python
plt.subplot(221)
plt.plot(df1['Actual'].iloc[:11],label='Actual')
plt.plot(df1['lr'].iloc[:11],label='lr')
plt.legend()

plt.subplot(222)
plt.plot(df1['Actual'].iloc[:11],label='Actual')
plt.plot(df1['svr'].iloc[:11],label='svr')
plt.legend()

plt.subplot(223)
plt.plot(df1['Actual'].iloc[:11],label='Actual')
plt.plot(df1['rfg'].iloc[:11],label='rfg')
plt.legend()

plt.subplot(224)
plt.plot(df1['Actual'].iloc[:11],label='Actual')
plt.plot(df1['gbr'].iloc[:11],label='gbr')
plt.legend()
```

# EVALUATE THE MODEL

```python
from sklearn import metrics
```

```python
#calculating r2 score 
score1=metrics.r2_score(y_test,y_pred1)
score2=metrics.r2_score(y_test,y_pred2)
score3=metrics.r2_score(y_test,y_pred3)
score4=metrics.r2_score(y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
#claculating mean absiolute error
s1=metrics.mean_absolute_error(y_test,y_pred1)
s2=metrics.mean_absolute_error(y_test,y_pred2)
s3=metrics.mean_absolute_error(y_test,y_pred3)
s4=metrics.mean_absolute_error(y_test,y_pred4)
```

```python
print(s1,s2,s3,s4)
```
