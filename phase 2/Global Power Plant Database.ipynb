# Filling Null values In 'latitude ' column

```python
#first check for outliers present in the same column

sns.boxplot(df.latitude)
```

```python
df.latitude=df['latitude'].fillna(df['latitude'].mean())
```

```python
df['latitude'].isnull().sum()
```

# Filling Null values In longitude column

```python
#checking for outliers using boxplot

sns.boxplot(df.longitude)
```

```python
#filling null values using interpolate linear method base on neighbour values
df['longitude']=df['longitude'].interpolate(method='linear')
```

```python
df['longitude'].isnull().sum()
```

# Removing the columns has maximum null values

```python
df=df.drop('other_fuel1',axis=1)
df=df.drop('other_fuel2',axis=1)
df=df.drop('other_fuel3',axis=1)
```

```python
df.isnull().sum()
```

# Filling null values in 'commissioning_year' column usif MOD values

```python
mode=df['commissioning_year'].mode()
df['commissioning_year']=df['commissioning_year'].fillna(mode[0])
```

```python
mode
```

# dropping the columns which has most null values

```python
#dropping 'wepp_id ,'generation_gwh_2013',estimated_generation_gwh' from dataset
df1=df.drop(['wepp_id','generation_gwh_2013','estimated_generation_gwh','generation_gwh_2019'],axis=1)
```

```python
df1
```

```python
#filling null values of owner column

mode=df1['owner'].mode()
df1['owner']=df1['owner'].fillna(mode[0])
```

```python
#filling null values of geolocation_source column

mode=df1['geolocation_source'].mode()
df1['geolocation_source']=df1['geolocation_source'].fillna(mode[0])
```

```python
#filling null values using mean values
df1['year_of_capacity_data']=df1['year_of_capacity_data'].fillna(df1['year_of_capacity_data'].mean())

df1['generation_gwh_2014']=df1['generation_gwh_2014'].fillna(df1['generation_gwh_2014'].mean())

df1['year_of_capacity_data']=df1['year_of_capacity_data'].fillna(df1['year_of_capacity_data'].mean())
```

```python
df1['generation_gwh_2015']=df1['generation_gwh_2015'].fillna(df1['generation_gwh_2015'].mean())
df1['generation_gwh_2016']=df1['generation_gwh_2016'].fillna(df1['generation_gwh_2016'].mean())
df1['generation_gwh_2017']=df1['generation_gwh_2017'].fillna(df1['generation_gwh_2017'].mean())
df1['generation_gwh_2018']=df1['generation_gwh_2018'].fillna(df1['generation_gwh_2018'].mean())
```

```python
mode=df1['generation_data_source'].mode()
df1['generation_data_source']=df1['generation_data_source'].fillna(mode[0])
```

```python
#dropping url column
df1=df1.drop('url',axis=1)
```

# Seperating Features into x and y

```python
x=df1.drop('primary_fuel',axis=1)
y=df1['primary_fuel']
```

# Feature Scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
```

# checking variance inflation fector

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif['vif values']=[variance_inflation_factor(x.values,i) for i in range(len(x.columns))]
vif['Features']=x.columns
vif
```

# importing train test split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

# import the models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)

svc=SVC()
svc.fit(x_train,y_train)

rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)

gbc=GradientBoostingClassifier()
gbc.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
y_pred2=svc.predict(x_test)
y_pred3=rfc.predict(x_test)
y_pred4=gbc.predict(x_test)
```

```python
models=pd.DataFrame({'Actual':y_test,'Lr':y_pred1,'svc':y_pred2,'rfc':y_pred3,'gbc':y_pred4})
```

```python
models
```
