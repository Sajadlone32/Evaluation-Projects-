# Encoding the traget variable using map method

```python
#first copy the dataset
df2=df.copy()
```

```python
df2['Attrition']=df2['Attrition'].map({'Yes':1,'No':0})
```

```python
df2['Attrition'].value_counts()
```

# DATA ENCODING USING ORDINAL ENCODER

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
```

```python
for i in df2.columns:
    if df2[i].dtype == 'object':
        rs=df2[i].values.reshape(-1,1)
        df2[i]=oe.fit_transform(rs)
df2.head()
```

```python
#checking the dataset info again
df2.info()
```

# spliting the dataset into x and y variable

```python
x=df2.drop('Attrition',axis=1)
y=df2['Attrition']
```

# creating training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

# import the models 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)

svc=SVC()
svc.fit(x_train,y_train)

rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)

gbc=GradientBoostingClassifier()
gbc.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
y_pred2=svc.predict(x_test)
y_pred3=rfc.predict(x_test)
y_pred4=gbc.predict(x_test)
```
